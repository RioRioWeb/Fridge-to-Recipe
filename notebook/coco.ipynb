```python
def crop_and_save_image(image_path, bbox, cate, id):
    # 画像を読み込む
    image = cv2.imread(image_path)
    if image is None:
        print("画像が読み込めませんでした。パスを確認してください。")
        return
    
    # バウンディングボックスの座標を整数に変換
    x, y, width, height = map(int, bbox)
    
    # バウンディングボックスに基づいて画像をクロップ
    cropped_image = image[y:y+height, x:x+width]
    output_path = DATA_DIR + 'data/' + cate + '/' + id + '.jpg'
    cv2.imwrite(output_path, cropped_image)
    # # クロップされた画像を表示
    # cv2.imshow('Cropped Image', cropped_image)
    # cv2.waitKey(0)  # 任意のキーが押されるまでウィンドウを開いたままにする
    # cv2.destroyAllWindows()  # ウィンドウを閉じる
```

```python
import os

def create_directory_if_not_exists(directory_path):
    # ディレクトリが存在しない場合にディレクトリを作成
    if not os.path.exists(directory_path):
        os.makedirs(directory_path)
        print(f"ディレクトリが作成されました: {directory_path}")
    else:
        print(f"ディレクトリはすでに存在します: {directory_path}")

# 使用例
# directory_path = 'path/to/your/directory'
# create_directory_if_not_exists(directory_path)
```

```python
image_id_dict = {id:item['file_name'] for id, item in enumerate(data['images'])}
cate_id_dict = {item['id']:item['name'] for item in data['categories']}
print(image_id_dict)
print(cate_id_dict)
for i, item in enumerate(data['annotations']):
    image_id = item['image_id'] # 画像id
    category_id = item['category_id'] # カテゴリid
    bbox = item['bbox'] # bboxのリスト
    directory_path = DATA_DIR + 'data/' + cate_id_dict[category_id]
    create_directory_if_not_exists(directory_path)
    # print(image_id_dict[image_id], cate_id_dict[category_id], bbox)
    # img = cv2.imread(DATA_DIR + image_id_dict[image_id])
    input_img_path = DATA_DIR + image_id_dict[image_id]
    crop_and_save_image(input_img_path, bbox, cate_id_dict[category_id], str(i))
    # img = img[:, :, [bbox[], 1, 0]]
    
    
```

```python
import cv2
import numpy as np

def crop_and_display_image(image_path, bbox):
    # 画像を読み込む
    image = cv2.imread(image_path)
    if image is None:
        print("画像が読み込めませんでした。パスを確認してください。")
        return
    
    # バウンディングボックスの座標を整数に変換
    x, y, width, height = map(int, bbox)
    
    # バウンディングボックスに基づいて画像をクロップ
    cropped_image = image[y:y+height, x:x+width]
    
    # クロップされた画像を表示
    cv2.imshow('Cropped Image', cropped_image)
    cv2.waitKey(0)  # 任意のキーが押されるまでウィンドウを開いたままにする
    cv2.destroyAllWindows()  # ウィンドウを閉じる

# 例: 画像パスとバウンディングボックスのデータ
image_path = DATA_DIR + 'DSC_5941_JPG_jpg.rf.7f34ef03affd2f952f6519e8506d8cdc.jpg'
bbox = [100, 50, 200, 150]  # 例えば x=100, y=50, width=200, height=150

crop_and_display_image(image_path, bbox)
```
